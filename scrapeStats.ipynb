```python

import re
import time
from difflib import get_close_matches
from itertools import islice

import pandas
import requests
from bs4 import BeautifulSoup
from openpyxl import load_workbook

start = time.time()
# assign url to variable and use requests to get html
print('Getting a list of all teams.')
url = 'https://www.teamrankings.com/nfl/teams/'
teams_page = requests.get(url)
teams = teams_page.content

# parse html with BeautifulSoup
soup = BeautifulSoup(teams, 'html.parser')

# scrap all team names and add to a dictionary
full_teams_list = {}
lst = soup.find_all('td', {'class': 'text-left'})

for item in lst:
    team_name = item.find('a').text
    # scrap url reference for team name (e.g., 'arizona-cardinals') and add to
    # dictionary
    url_ref = item.find('a')['href'][10:]
    full_teams_list[team_name] = url_ref
print('Done!')

def take(n, iterable):
    """Return first n items of the iterable as a list."""
    return list(islice(iterable, n))

# ugly but cant figure out why it's not scraping properly
del full_teams_list['Rankings']


#create list we will use below to grab stats
teams_list = []
url_ref_list = []

for teams in full_teams_list.items():
    teams_list.append(teams[0])
    url_ref_list.append(teams[1])
    



```

```python
## second part..

# build two lists of urls to scrap stats for each team
print('Building two lists of urls to parse')
stat_url_list = []
for url in url_ref_list:
    stat_url_list.append('https://www.teamrankings.com/nfl/team/' +
                         url + '/stats')
print('Done!')

print('Getting stats from teamrankings.com')
# create empty lists to hold stats
ypp = []
oYPP = []
qbsacked = []
sackpct = []

for stat in range(len(stat_url_list)):
    stats_page = requests.get(stat_url_list[stat])
    
    stats = stats_page.content

    soup1 = BeautifulSoup(stats, 'html.parser')


    numbers = soup1.find_all('td', {'class': 'nowrap'})
    s = []

    for number in numbers:
        s.append(number.text)
        
        
    try:
        ypp.append(float(s[13][:4]))
    except:
        ypp.append(0.00)
        
    try:
        oYPP.append(float(s[15][:4]))
    except:
        oYPP.append(0.00)
        
    try:
        qbsacked.append(float(s[77][:4]))
    except:
        qbsacked.append(0.00)
        
    try:
        sackpct.append(float(s[79][:4]))
    except:
        sackpct.append(0.00)
```

```python
 def team_stats():
        offtable = {}
        offlist = []
        for i in range(len(full_teams_list)):
            team = teams_list[i]
            ypp_stat = ypp[i]
            oYPP_stat = oYPP[i]
            qbsacked_stat = qbsacked[i]
            sackpct_stat = sackpct[i]
            offtable.update (teamname = team, yppcol = ypp_stat, oYPPcol = oYPP_stat, qbsackedcol = qbsacked_stat, sackpctcol = sackpct_stat)
            offlist.append(dict(offtable))
            df = pandas.DataFrame(offlist)
        return df
    
    
    

```
