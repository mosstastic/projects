```python

import pandas as pd
import pandas.io
from bs4 import BeautifulSoup
import urllib3
import requests

```

```python
def FO():
    def convert_DVOA_table(theTable):
        tableList = []
        for tr in theTable.find_all('tr'):
            tds = tr.find_all('td')
            if tds == []:
                continue
            tableRow = []
            for thisCell in tds:
                try:
                    tableRow.append(float(thisCell.text.replace('%','')))
                except:
                    tableRow.append(thisCell.text.replace('%',''))
            if tableRow[0] != '':
                tableList.append(tableRow)
        return tableList
    
    

    def ann_off_data(annumrange):
        offtable = {}
        for annum in annumrange:
            tableurl = "http://www.footballoutsiders.com/stats/nfl/team-efficiency/" + str(annum)
            teams_page = requests.get(tableurl)
            teams = teams_page.content
            soup = BeautifulSoup(teams, 'html.parser')
            stattable = soup.find("table", class_ = "stats")
            statarray = convert_DVOA_table(stattable)
            offtable[annum] = pd.DataFrame(statarray) #, columns=cols)
        return offtable

    offense = ann_off_data(range(2020,2021))
    offense[2020]
    
    translation = {'ARI': 'Arizona Cardinals','ATL': 'Atlanta Falcons','BAL': 'Baltimore Ravens','BUF': 'Buffalo Bills','CAR': 'Carolina Panthers','CHI': 'Chicago Bears','CIN': 'Cincinnati Bengals','CLE': 'Cleveland Browns','DAL': 'Dallas Cowboys','DEN': 'Denver Broncos','DET': 'Detroit Lions','GB': 'Green Bay Packers','HOU': 'Houston Texans','IND': 'Indianapolis Colts','JAX': 'Jacksonville Jaguars','KC': 'Kansas City Chiefs','MIA': 'Miami Dolphins','MIN': 'Minnesota Vikings','NE': 'New England Patriots','NO': 'New Orleans Saints','NYG': 'New York Giants','NYJ': 'New York Jets','LV': 'Las Vegas Raiders','PHI': 'Philadelphia Eagles','PIT': 'Pittsburgh Steelers','LAC': 'Los Angeles Chargers','SF': 'San Francisco 49ers','SEA': 'Seattle Seahawks','LAR': 'Los Angeles Rams','TB': 'Tampa Bay Buccaneers','TEN': 'Tennessee Titans'}
    df2 = pd.DataFrame(list(translation.items()),columns=['Short','Long'])


    cols = ['Overall Rank','Team','Total DVOA',\
                        'Last Week','Total Dave',\
                        'Rank','W-L',\
                        'Offense DVOA','OFF Rank',\
                        'Defense DVOA','DEF Rank','ST DVOA',\
                        'ST Rank'
                       ]


    #create dataframe 
    output = pd.DataFrame(offense[2020])

    #assign column names
    output.columns = cols
    finaloutput = pd.merge(output, df2, left_on = "Team", right_on = "Short")

    return finaloutput.iloc[:,[1,7,9,-1]]
```
